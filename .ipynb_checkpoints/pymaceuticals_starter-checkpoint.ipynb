## Observations and Insights 

* There is a slightly higher percentage (1%) of male mice.
* There is a positive relationship between weight and average tumor volume. The correlation coefficient is 0.95.
* Mouse ID: s185 treated with Capomulin has seen a gradual decrease of tumor volume over time.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata,study_results, on="Mouse ID", how = "outer")
combined_df
```

```python
# Checking the number of mice in the DataFrame.
numberofmice = len(combined_df["Mouse ID"].unique())
numberofmice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_df = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_df_data = combined_df.loc[combined_df["Mouse ID"]=="g989",["Mouse ID","Drug Regimen", "Sex", "Age_months", "Weight (g)","Timepoint", "Tumor Volume (mm3)","Metastatic Sites"]]
duplicate_df_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
index = combined_df[combined_df["Mouse ID"] == 'g989'].index
combined_df.drop(index, inplace=True)
combined_df
```

```python
# Checking the number of mice in the clean DataFrame.
updatednumbermice = len(combined_df["Mouse ID"].unique())
updatednumbermice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
grouped_regimen_df = combined_df.groupby(['Drug Regimen'])


#Calculate mean
cmean = grouped_regimen_df["Tumor Volume (mm3)"].mean()

#Calculate median
cmedian = grouped_regimen_df["Tumor Volume (mm3)"].median()

#Calculate variance
cvar = grouped_regimen_df["Tumor Volume (mm3)"].var()

#Calculate standard deviation
cstd = grouped_regimen_df["Tumor Volume (mm3)"].std()

#Calculate SEM
csem = grouped_regimen_df["Tumor Volume (mm3)"].sem()

# This method is the most straighforward, creating multiple series and putting them all together at the end.
summarystatistics_df = pd.DataFrame({"Mean": cmean, "Median": cmedian,"Variance":cvar,"Standard Deviation": cstd,"SEM": csem})
summarystatistics_df.sort_values("Median",ascending = False)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

new_groupby = combined_df.groupby(["Mouse ID","Drug Regimen"])
#new_combined_df1 = new_combined_df.loc[new_combined_df["Drug Regimen"]== ("Capomulin" and "Ramicane" and "Infubinol" and "Ceftamin"),:] 

groupby_mean = new_groupby["Tumor Volume (mm3)"].mean()

groupby_mean_df = pd.DataFrame({"Avg Tumor Volume":groupby_mean})
groupby_mean_df = groupby_mean_df.reset_index()

#top four Drug Regimen
topfour = ["Capomulin","Ramicane","Infubinol","Ceftamin"]
topDR_df = groupby_mean_df.loc[groupby_mean_df["Drug Regimen"].isin(topfour)]
 
#calculate quartiles
quartiles = topDR_df["Avg Tumor Volume"].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of occupancy is: {round(lowerq,2)}")
print(f"The upper quartile of occupancy is: {round(upperq,2)}")
print(f"The interquartile range of occupancy is: {round(iqr,2)}")
print(f"The the median of occupancy is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"Values below {round(lower_bound,2)} are possibly outliers.")
print(f"Values above {round(upper_bound,2)} are possibly outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

topDR_df

Ceftamin = []
Infubinol = []
Ramicane = []
Capomulin = []

for index,row in topDR_df.iterrows():
    if row["Drug Regimen"] == "Ceftamin":
        Ceftamin.append(row["Avg Tumor Volume"])
    if row["Drug Regimen"] == "Infubinol":
        Infubinol.append(row["Avg Tumor Volume"])
    if row["Drug Regimen"] == "Ramicane":
        Ramicane.append(row["Avg Tumor Volume"])
    if row["Drug Regimen"] == "Capomulin":
        Capomulin.append(row["Avg Tumor Volume"])
    
topfour_df = pd.DataFrame({"Capomulin":Capomulin,"Ramicane":Ramicane,"Infubinol":Infubinol,"Ceftamin":Ceftamin})
topfour_df.plot.box()
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
lineplot_df = combined_df.loc[combined_df["Mouse ID"]=="s185",["Mouse ID","Drug Regimen", "Sex", "Age_months", "Weight (g)","Timepoint", "Tumor Volume (mm3)","Metastatic Sites"]]
Timepoint = lineplot_df["Timepoint"].to_list()
Tumorvolume = lineplot_df["Tumor Volume (mm3)"].to_list()
plt.plot(Timepoint,Tumorvolume)
plt.title("Capomulin: Mouse s185")
plt.xlabel("Time Point")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin = combined_df.loc[combined_df["Drug Regimen"] == "Capomulin", ["Weight (g)","Tumor Volume (mm3)"]]
capomulin_df = pd.DataFrame(capomulin)
capomulin_grouped = capomulin_df.groupby(["Weight (g)"])

Tumorvolumeavg = capomulin_grouped.mean()
Tumorvolumeavg_df = pd.DataFrame(Tumorvolumeavg)

#add index numbers
Tumorvolumeavg_df["Index"] = range(0,0+len(Tumorvolumeavg_df))
Tumorvolumeavg_df["Weight"] = Tumorvolumeavg_df.index
Tumorvolumeavg_df1 = Tumorvolumeavg_df.set_index("Index")

#convert to lists
avgtumor = Tumorvolumeavg_df1["Tumor Volume (mm3)"]
weight = Tumorvolumeavg_df1["Weight"]

plt.scatter(weight, avgtumor, marker="o", facecolors="red", edgecolors="black")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Capomulin: Mouse Weight vs Avg Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

y = st.pearsonr(weight,avgtumor)
print(f'The correlation coefficient is {round(st.pearsonr(weight,avgtumor)[0],2)}.')

#Perform linear regression
c_slope, c_int, c_r, c_p, c_std_err = st.linregress(weight,avgtumor)

#Create equation
c_fit = c_slope * weight + c_int

plt.scatter(weight, avgtumor, marker="o", facecolors="red", edgecolors="black")
plt.plot(weight,c_fit,"--")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Capomulin: Mouse Weight vs Avg Tumor Volume")
plt.show()
```
